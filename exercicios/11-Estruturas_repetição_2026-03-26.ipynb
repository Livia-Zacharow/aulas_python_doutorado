Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

* R.: 1 - percorrer todos os valores da lista e, para cada um deles:

2 - dividir por 2 e analisar o resto da divisão:

3a - se o resto da divisão (%) for igual a zero, entao o valor é par

3b - se o resto da divisão for qualquer outro valor, então é ímpar

## 03 – Escreva o programa em Python:

```python
lista = list(range(5000))

pares = 0
impares = 0

for numero in lista:
    if numero % 2 == 0:
        pares +=1
    else:
        impares +=1
print("Pares:", pares)
print("Ímpares:", impares)
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
lista = list(range(5000))
n = len(lista)

par = n // 2
impar = n // 2

print(par)
print(impar)
```
