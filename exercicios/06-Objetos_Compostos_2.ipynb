## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
entradas = len(df_ex10)
```

```python
tipos_colunas = df_ex10.dtypes.tolist()
```

```python
contagem_tipos = df_ex10.dtypes.value_counts().tolist()
```

```python
lista_1 = df_ex10.iloc[:, 0]
lista_2 = df_ex10.iloc[:, 3]
lista_3 = df_ex10.iloc[:, 8]

print(lista_1)
print(lista_2)
print(lista_3)
```

```python
medias = [lista_1.mean(),
          lista_2.mean(),
          lista_3.mean()]

medias
```

```python
desvios = [lista_1.std(),
           lista_2.std(),
           lista_3.std()]

desvios
```

```python
terceiro_quartil = [lista_1.quantile(0.75),
                    lista_2.quantile(0.75),
                    lista_3.quantile(0.75)]

terceiro_quartil
```
