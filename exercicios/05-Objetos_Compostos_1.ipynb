```python
#Lista interleucina_12#
interleucina_12_lista = [
    len(genes_imuno[3]),
    genes_imuno[3][:2],
    genes_imuno[3].count("a"),
    "hu" in genes_imuno[3]
]
print(interleucina_12_lista)
```

```python
#Lista TGF_beta#
TGF_beta_lista = [
    len(genes_imuno[4]),
    genes_imuno[4][:2],
    genes_imuno[4].count("a"),
    "hu" in genes_imuno[4]
]
print(TGF_beta_lista)
```
